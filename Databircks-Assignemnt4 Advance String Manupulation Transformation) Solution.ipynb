#### Importing Libraries

```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
MySchema = StructType([StructField("Id", IntegerType(),False),
                     StructField("first_name", StringType(),True),
                     StructField("last_name", StringType(),True),
                     StructField("email", StringType(),True),
                     StructField("gender", StringType(),True),
                     StructField("salary", DecimalType(),True),
                     StructField("creationDate", StringType(),True),
                     StructField("bonus", DecimalType(),True)
                ]
)
```

```
df = spark.read.csv("/FileStore/tables/EmployeeData-1.csv", header=True,schema=MySchema)
df.printSchema()
display(df)
```

<b> What is the abbreviated version of each employee's last name, with the first letter and the two following letters included?

```
# Write your code using the Spark Function
df.select("last_name",substring("last_name",1,3).alias("Abbreviated")).show()
```

d
<b> What is the domain name of each employee's email address, so that their emails can be filtered and routed to the correct department?

```
# Write your code using the Spark Function
df = df.withColumn("Employee",concat_ws(' ',df.first_name,df.last_name))
df.filter("email is not null").select("Employee",split(df['email'],"@")[1].alias("Domain")).show()
```

d
<b> In an effort to protect their privacy, what is each employee's name spelled backwards?

```
# Write your code using the Spark Function
df.select(col("Employee"),reverse(col("Employee")).alias("Reverse Employee")).show()
```

d
<b> In order to make their email addresses harder to guess, what is each employee's email address spelled backwards?

```
# Write your code using the Spark Function
df.select(col("email"),reverse(col("email")).alias("Reverse Email")).show()
```

d
<b> What is each employee's first name, repeated twice, to improve personalization of emails and messages?

```
# Write your code using the Spark Function
df.groupby("first_name").agg(count("first_name").alias("Cnt")).filter("Cnt > 1").show()
```
