#### Importing Libraries

```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
MySchema = StructType([StructField("Id", IntegerType(),False),
                     StructField("first_name", StringType(),True),
                     StructField("last_name", StringType(),True),
                     StructField("email", StringType(),True),
                     StructField("gender", StringType(),True),
                     StructField("salary", DecimalType(),True),
                     StructField("creationDate", StringType(),True),
                     StructField("bonus", DecimalType(),True)
                ]
)
```

```
df = spark.read.csv("/FileStore/tables/EmployeeData-1.csv", header=True,schema=MySchema)
df.printSchema()
display(df)
```

<b> How many employees have a first name that starts with the letter 'J'?

```
# Write your code using the Spark Function
df.filter('first_name like "J%"').show()
```

d
<b> How many employees have a last name that ends with the letter 's'?

```
# Write your code using the Spark Function
df.filter('last_name like "%s"').agg(count("Id")).show()
```

d
<b> What is the length of the longest first name?

```
# Write your code using the Spark Function
df1 = df.withColumn('length',length('first_name'))
df1.filter("length is not null").orderBy(col("length").desc()).show(1)
```

d
<b> What is the length of the shortest last name?

```
# Write your code using the Spark Function
df1.filter("length is not null").orderBy(col("length").asc()).show(1)
```

d
<b> What is the email domain of the employee with the highest salary?

```
# Write your code using the Spark Function
df1.filter("email is not null").select("salary",split(df1['email'],"@")[1].alias("Domain")).orderBy(col("salary").desc()).show()
```

d
<b> What is the first name of the employee with the email address that contains the substring 'hr'?

```
# Write your code using the Spark Function
df.filter('first_name like "%hr%"').select("first_name","email").show()
```

d
<b> What is the last name of the employee with the email address that ends with the domain '.com'?

```
# Write your code using the Spark Function
display(df.filter('email like "%.com"').select("last_name","email"))
```

d
<b> How many employees have an email address that contains their last name?

```
# Write your code using the Spark Function
display(df.filter('last_name is not null').select("last_name","email"))
```

d
<b> What is the email of the employee with the first name that contains a space character?

```
# Write your code using the Spark Function
df.filter('first_name like "% %"').select("first_name","email").show()
```

d
<b> What is the email of the employee with the last name that contains a hyphen character?

```
# Write your code using the Spark Function
df.filter('last_name like "%-%"').select("last_name","email").show()
```
